```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
time_train = pd.read_csv('./data/Train.csv')
time_test = pd.read_csv('./data/Test.csv')
```

```python
time_train
```

```python
plt.plot(time_train['Count'])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

series = time_train['Count']

# seasonal decompose with period = 24 (1 day)

result = seasonal_decompose(series, model='additive', period=24)
result.plot()
plt.show()
```

```python
# seasonal decompose with period = 24*7 (1 week)

result = seasonal_decompose(series, model='additive', period=24*7)
result.plot()
plt.show()
```

```python
# seasonal decompose with period = 24*30 (1 month)

result = seasonal_decompose(series, model='additive', period=24*30)
result.plot()
plt.show()
```

```python
# seasonal decompose with period = 24*30*6 (6 month)

result = seasonal_decompose(series, model='additive', period=24*30*6)
result.plot()
plt.show()
```
