```python
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
import pandas as pd
import numpy as np
import torch
from sklearn.metrics.pairwise import cosine_similarity
from transformers import AutoTokenizer, AutoModel
```

```python
df = pd.read_csv('dailyo.csv')
dff = df.copy()
dff['News'] = dff['News_title'] + " " + dff['News_headline']
dff.dropna(inplace=True)
```

```python
dff['News'].isna().sum()
dfff = dff['News'].dropna()
```

```python
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

all_news = list(dfff)
out_embs = []

for n in tqdm(all_news):
    embd = model.encode(n)
    out_embs.append(embd)
out_embs = np.array(out_embs)
out_embs.shape
```

```python
query = "Chai Vala"
qy_emb = model.encode([query])
sims = []
for e in tqdm(out_embs):
    sims.append(model.similarity(qy_emb,e).numpy()[0])
df_ind = dff[['News']]
df_ind['sims'] = np.array(sims)
df_ind.sort_values('sims', ascending=False)
```
