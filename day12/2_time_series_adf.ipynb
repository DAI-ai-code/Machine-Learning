## Augumented Dicky Fuller

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose 
from statsmodels.tsa.api import adfuller
```

```python
df = pd.read_csv('../Datasets/FRED-NROUST.csv')
y = df['Value']
y.plot()
```

```python
res = adfuller(y, maxlag=10)
res
```

<h4 style='color: green'>2nd value in above tuple is p-value</h4>

```python
p_value = res[1]
if p_value < 0.05:
    print("stationary")
else:
    print("Not stationary")
```

```python
ord_1_diff = y.diff()
ord_1_diff = ord_1_diff.dropna()

res = adfuller(ord_1_diff, maxlag=10)
p_value = res[1]
if p_value < 0.05:
    print('stationary ->', p_value)
else:
    print('Not stationary ->', p_value)


```
