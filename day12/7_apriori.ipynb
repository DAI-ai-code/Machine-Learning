```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
fp_df = pd.read_csv('../Datasets/Faceplate.csv', index_col = 0)
fp_df.head()
```

```python
fp_df = fp_df.astype(bool)
itemsets = apriori(fp_df, min_support=0.2, use_colnames=True)
rules = association_rules(itemsets, metric='confidence', min_threshold=0.5)
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules
```

```python
from ipywidgets import widgets

def ass_rule(supp, thresh):
    fp_df = pd.read_csv('../Datasets/Faceplate.csv', index_col = 0)
    fp_df = fp_df.astype(bool)
    itemsets = apriori(fp_df, min_support=supp, use_colnames=True)
    rules = association_rules(itemsets, metric='confidence', min_threshold=thresh)
    rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
    return rules
```

```python
widgets.interact(ass_rule, supp=(0.01, 1, 0.01), thresh=(0.01, 1, 0.01))
```
