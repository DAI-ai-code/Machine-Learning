```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Datasets/milk.csv', index_col=0)

rif = IsolationForest(contamination=0.05, random_state=25)
rif.fit(df)
predictions = rif.predict(df)
(dict(zip(list(df.index),list(predictions))))
```

```python
df = pd.read_csv('../Datasets/nutrient.csv', index_col=0)

from sklearn.ensemble import IsolationForest

iso_for = IsolationForest(random_state=25, contamination=0.05)
iso_for.fit(df)
pred = iso_for.predict(df)

list(zip(df.index, pred))
```

without contamination input

```python
df = pd.read_csv('../Datasets/nutrient.csv', index_col=0)

from sklearn.ensemble import IsolationForest

iso_for = IsolationForest(random_state=25)
iso_for.fit(df)
pred = iso_for.predict(df)

dict(zip(df.index, map(int, pred)))
```

used map(int, pred) to convert np.int64 numbers to simple int dtypes.
