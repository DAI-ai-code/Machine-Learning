```python
import numpy as np
import pandas as pd
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
```

```python
comp_prob = pd.read_csv('../Datasets/comp_prob.csv')
```

```python
y_pred_probs1 = comp_prob['yprob_1']
y_pred_probs2 = comp_prob['yprob_2']
y_test = comp_prob['y_test']
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_probs1)
plt.plot([0, 1], [0, 1])
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.tight_layout()
plt.show()
```

```python
roc_auc_score(y_test, y_pred_probs1)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_probs2)
plt.plot([0, 1], [0, 1])
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.tight_layout()
plt.show()
```

```python
roc_auc_score(y_test, y_pred_probs2)
```
