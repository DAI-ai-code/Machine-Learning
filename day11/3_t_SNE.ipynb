```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder, MinMaxScaler
from sklearn.compose import make_column_transformer, make_column_selector
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.manifold import TSNE
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('../Datasets/cases/Wisconsin/BreastCancer.csv', index_col=0)
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
tsne = TSNE(random_state=25).set_output(transform='pandas')
X_tsne = tsne.fit_transform(X)
tsne_data = X_tsne.copy()
tsne_data['Class'] = y
sns.scatterplot(data=tsne_data, x='tsne0', y='tsne1', hue='Class')
```

```python
pca = PCA().set_output(transform='pandas')

X_pca = pca.fit_transform(X)

X.shape, X_pca.shape
X_pca_copy = X_pca.copy()
X_pca_copy['Class'] = y

import seaborn as sns

sns.scatterplot(data=X_pca_copy, x='pca0', y='pca1', hue='Class')
```

```python
df = pd.read_csv('../Datasets/cases/Vehicle_Silhouettes/Vehicle.csv')
std = StandardScaler().set_output(transform='pandas')
X = df.drop('Class', axis=1)
y = df['Class']
```

without scaling

PCA

```python


prcomp = PCA().set_output(transform='pandas')
X_pca = prcomp.fit_transform(X)
X_pca['Class'] = y
sns.scatterplot(data=X_pca, x='pca0', y='pca1', hue='Class')
```

tSNE

```python
tsne = TSNE(perplexity=100,random_state=25).set_output(transform='pandas')
X_tsne = tsne.fit_transform(X)
X_tsne['Class'] = y
sns.scatterplot(data=X_tsne, x='tsne0', y='tsne1', hue='Class')
```

with scaling

```python
X_scl = std.fit_transform(X)
```

PCA

```python
prcomp = PCA().set_output(transform='pandas')
X_pca = prcomp.fit_transform(X_scl)
X_pca['Class'] = y
sns.scatterplot(data=X_pca, x='pca0', y='pca1', hue='Class')
```

tSNE

```python
tsne = TSNE(perplexity=2,random_state=25).set_output(transform='pandas')
X_tsne = tsne.fit_transform(X_scl)
X_tsne['Class'] = y
sns.scatterplot(data=X_tsne, x='tsne0', y='tsne1', hue='Class')
```
