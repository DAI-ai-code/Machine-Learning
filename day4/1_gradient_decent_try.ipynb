```python
xi = df['satisfaction_level']
yi = df['left']
n = 0.1

# initalize w and b
w = 0.1
b = 0.1
```

```python
import numpy as np
import pandas as pd

# Assuming df is your DataFrame with 'satisfaction_level' and 'left' columns
xi = df['satisfaction_level'].values
yi = df['left'].values
n = 0.1  # learning rate

# Initialize w and b
w = 0.1
b = 0.1

# Number of iterations for gradient descent
epochs = 1000

def sigmoid(z):
    """Sigmoid activation function"""
    return 1 / (1 + np.exp(-z))

def compute_loss(y_true, y_pred):
    """Binary cross-entropy loss"""
    epsilon = 1e-15  # to avoid log(0)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# Gradient descent
for epoch in range(epochs):
    # Forward pass
    z = w * xi + b
    y_pred = sigmoid(z)
    
    # Compute loss (optional - for monitoring)
    loss = compute_loss(yi, y_pred)
    
    # Compute gradients
    # dabba L / dabba w = (1/m) * sum((y_pred - y_true) * xi)
    # dabba L / dabba b = (1/m) * sum(y_pred - y_true)
    
    dw = np.mean((y_pred - yi) * xi)  # âˆ‚L/âˆ‚w
    db = np.mean(y_pred - yi)         # âˆ‚L/âˆ‚b
    
    # Update parameters
    w = w - n * dw
    b = b - n * db
    
    # Print progress every 100 epochs
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.4f}, w = {w:.4f}, b = {b:.4f}")

print(f"\nFinal parameters: w = {w:.4f}, b = {b:.4f}")

# Make predictions
def predict(x, w, b):
    """Make predictions using the trained model"""
    z = w * x + b
    return (sigmoid(z) > 0.5).astype(int)

# Test the model
predictions = predict(xi, w, b)
accuracy = np.mean(predictions == yi)
print(f"Accuracy: {accuracy:.4f}")
```
