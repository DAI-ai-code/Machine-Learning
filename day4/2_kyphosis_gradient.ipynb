```python
import pandas as pd
import numpy as np
from sklearn.metrics import log_loss, classification_report
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
kyph = pd.read_csv('../Datasets/cases/Kyphosis/Kyphosis.csv')
le = LabelEncoder()
kyph.head()
```

```python
kyph['Kyphosis'] = le.fit_transform(kyph['Kyphosis'])
y = kyph['Kyphosis']
X = kyph.drop('Kyphosis', axis=1)
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=25)
```

```python
mm = MinMaxScaler()
X_trn_scl = mm.fit_transform(X_train)
X_tst_scl = mm.fit_transform(X_test)
```

```python
eta = 0.4
b,w1,w2,w3 = 0.1,0.1,0.1,0.1
epoch = 10000
# x1 = X_trn_scl['Age']
# x2 =  X_trn_scl['Number']
# x3 = X_trn_scl['Start']

# Use integer indexing since X_trn_scl is now a NumPy array
x1 = X_trn_scl[:, 0]  # First column (Age)
x2 = X_trn_scl[:, 1]  # Second column (Number)  
x3 = X_trn_scl[:, 2]  # Third column (Start)


scores = []
def sigmoid(z):
    return 1/(1+np.exp(-z))


for i in range(epoch):
    y_cap = sigmoid(w1*x1 + w2*x2 + w3*x3 + b)
    dw1 = -np.mean((y_train-y_cap)*x1)
    dw2 = -np.mean((y_train-y_cap)*x2)
    dw3 = -np.mean((y_train-y_cap)*x3)
    db = -np.mean(y_train - y_cap)

    L = log_loss(y_train,y_cap)
    scores.append([w1,w2,w3, b,L])
    w1 = w1 - eta*dw1
    w2 = w2 - eta*dw2
    w3 = w3 - eta*dw3
    b = b - eta*db

scores =  pd.DataFrame(scores, columns=['W1','W2','W3','B','Log Loss'])
scores.sort_values('Log Loss')
```

```python
x1 = X_tst_scl[:, 0]  # First column (Age)
x2 = X_tst_scl[:, 1]  # Second column (Number)  
x3 = X_tst_scl[:, 2]
y_pred_prob = sigmoid(w1*x1 + w2*x2 + w3*x3 + b)
print("Log loss on test: ", log_loss(y_test,y_pred_prob))
```

```python
y_pred = np.where(y_pred_prob>0.5,1,0)
print("classification report on test: ", classification_report(y_test,y_pred))
```
