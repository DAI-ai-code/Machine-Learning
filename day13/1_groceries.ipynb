We need to use basic python file management

```python
groceries = []

# opening file with with

with open('../Datasets/groceries.csv', 'r') as f:
    groceries = f.read()

groceries = groceries.split('\n')
groceries[:3]
```

```python
type(groceries)
```

```python
groceries_list = []
for i in groceries:
    groceries_list.append(i.split(','))
```

```python
# mlxtend imports

from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder().set_output(transform='pandas')
te_ary = te.fit_transform(groceries_list)
te_ary
```

```python
itemsets = apriori(te_ary, min_support=0.005, use_colnames=True)
rules = association_rules(itemsets, metric='confidence', min_threshold=0.005)
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules.sort_values(['lift', 'confidence'], ascending=[False, False])
```

```python
def create_rules(min_supp, conf_thres):
    itemsets = apriori(te_ary, min_support=min_supp, use_colnames=True)
    rules = association_rules(itemsets, metric='confidence', min_threshold=conf_thres)
    rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
    rules.sort_values(['lift', 'confidence'], ascending=False, inplace=True)
    return rules
    
```

```python
# widget import
from ipywidgets import widgets
```

```python
widgets.interact(create_rules, min_supp=(0.001, 1, 0.001), conf_thres=(0.001, 1, 0.001))
```
