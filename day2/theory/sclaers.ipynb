```python
import numpy as np
import pandas as pd

np.random.seed(4)
a = np.random.randint(0, 100, (100,3))
a[:, 1:2] *= 9

# Convert only the third column to float
third_col = a[:, 2].astype(float) * 0.1
df = pd.DataFrame({
    'X1': a[:, 0],
    'X2': a[:, 1], 
    'X3': third_col
})
print(df)
```
