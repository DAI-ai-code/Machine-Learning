```python
import pandas as pd
import numpy as np
from sklearn.metrics import log_loss, classification_report, f1_score
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
wisconsin = pd.read_csv("../Datasets/cases/Wisconsin/BreastCancer.csv", index_col=0)
```

```python
from sklearn.pipeline import Pipeline

X = wisconsin.drop('Class', axis=1)
y = wisconsin['Class']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=25)

svm = SVC()

scores = []
Cs = np.linspace(1e-3, 5, 20)
for c in Cs:
    pipe = Pipeline(
    [('SVM',SVC(kernel='linear', C=c))])
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)

    scores.append([c,f1_score(y_test,y_pred, pos_label='Malignant')])

scores = pd.DataFrame(scores, columns=['C','F1'])
scores.sort_values('F1', ascending=False)


```
