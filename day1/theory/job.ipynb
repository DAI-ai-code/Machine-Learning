```python
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
jobs = pd.read_csv('../../Datasets/JobSalary2.csv', index_col='S_No')
```

```python
jobs.isnull().sum()
```

```python
jobs
```

```python
imputer = SimpleImputer(strategy='mean').set_output(transform='pandas')
imputer.fit(jobs) # compute mean of each column
imputer.statistics_ # gives calculated means
imputer.transform(jobs) # fills missing value by the computed mean
```

```python
imputer = SimpleImputer(strategy='median').set_output(transform='pandas')
imputer.fit(jobs) # compute mean of each column
imputer.statistics_ # gives calculated means
imputer.transform(jobs) # fills missing value by the computed mean
```
